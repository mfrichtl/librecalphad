```python
from libreCalphad.databases.db_utils import load_database
import matplotlib.pyplot as plt
from pycalphad import Database, calculate
import pycalphad.variables as v

db = load_database('mf-steel.tdb')
components = ['FE', 'NI', 'VA']
phases = ['BCC_A2', 'FCC_A1', 'AL3M_D022', 'LIQUID']

fig,ax = plt.subplots()

for phase in phases:
    calc = calculate(db, components, phase, N=1, P=101325, T=900)
    ax.scatter(calc.sel(component='NI').X, calc.GM, label=phase)

ax.set_xlabel('NI')
fig.legend()
```
